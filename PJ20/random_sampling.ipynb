```python
import os
import shutil
import random

def copy_random_images(src_dir, dest_dir, num_images=150):
    # Ensure destination directory exists
    os.makedirs(dest_dir, exist_ok=True)
    
    # Get list of all batch directories in the source directory
    batch_dirs = [d for d in os.listdir(src_dir) if os.path.isdir(os.path.join(src_dir, d)) and d.startswith('batch')]
    
    for batch_dir in batch_dirs:
        batch_path = os.path.join(src_dir, batch_dir)
        images = [f for f in os.listdir(batch_path) if os.path.isfile(os.path.join(batch_path, f))]
        
        # Randomly select num_images from the list of images
        selected_images = random.sample(images, min(num_images, len(images)))
        
        for image in selected_images:
            src_image_path = os.path.join(batch_path, image)
            dest_image_path = os.path.join(dest_dir, f"{batch_dir}_{image}")
            shutil.copy(src_image_path, dest_image_path)
            # print(f"Copied {src_image_path} to {dest_image_path}")
        print(f"batch {batch_dir} Done")

if __name__ == "__main__":
    src_directory = 'D:/nsk/data/CH03_output'
    dest_directory = 'D:/nsk/data/CH03_train'
    
    copy_random_images(src_directory, dest_directory)
```
